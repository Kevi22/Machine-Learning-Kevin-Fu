## Cardiovasual diease report


### Read in file and handle missing values.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

filepath = '../Lab/data/cardio_train.csv'
df = pd.read_csv(filepath, delimiter=";")
df = df.dropna()
```

### Review the dataset

```python
df
```

```python
positive = df[df['cardio'] == 1].shape[0]
negative = df[df['cardio'] == 0].shape[0]
labels = ['Positive', 'Negative']
sizes = [positive, negative]
colors = ['#ff6361', '#2f4b7c']

print("Positive with cardiovascular diesease:",positive)
print("Negative with cardiovascular diesease:",negative)

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Cardiovascular Disease')
plt.axis('equal') 
plt.show()
```

```python
normal = df[df['cholesterol'] == 1].shape[0]
above = df[df['cholesterol'] == 2].shape[0]
high = df[df['cholesterol'] == 3].shape[0]
total = len(df)

print("Total people in the dataset:", total)
print("People with normal cholesterol level:",normal)
print("People with above normal cholesterol level:",above)
print("People with well above normal cholesterol level:",high)
```

```python
cholesterol_levels = [normal, above, high]
labels = ['Normal', 'Above Normal', 'Well Above Normal']
colors = ['#2f4b7c', '#a05195', '#ff6361']

plt.figure(figsize=(8, 6))
plt.bar(labels, cholesterol_levels, color=colors)
plt.xlabel('Cholesterol Levels')
plt.ylabel('Number of People')
plt.title('Distribution of Cholesterol Levels')
plt.show()
```

```python
total = len(df)
male = df[df['gender'] == 2].shape[0]
female = df[df['gender'] == 1].shape[0]

male_percentage = (round((male / total) * 100))
female_percentage = (round((female / total) * 100))

labels = ['Male', 'Female']
sizes = [male_percentage, female_percentage]
colors = ['#ff6361', '#2f4b7c']
explode = (0.1, 0)  # explode 1st slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.0f%%', startangle=140)
plt.axis('equal') 
plt.title('Gender Distribution')
plt.show()
```

```python
total = len(df)
smoker = df[df['smoke'] == 1].shape[0]
non_smoker = df[df['smoke'] == 0].shape[0]

smoker_percentage = (round((smoker / total) * 100))
non_smoker_percentage = (round((non_smoker / total) * 100))

labels = ['Smoker', 'Non-Smoker']
sizes = [smoker_percentage, non_smoker_percentage]
colors = ['#ff6361', '#2f4b7c']
explode = (0.1, 0)  

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.0f%%', startangle=140)
plt.axis('equal') 
plt.title('Smoking Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='height', hue='gender', bins=30, palette=['#ff6361', 'blue'])
plt.title('Distribution of Heights by Gender')
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.legend(labels=['Male', 'Female'])
plt.grid(True)
plt.show()
```

```python
cardio_positive = df[df['cardio'] == 1]

positive_male = cardio_positive[cardio_positive['gender'] == 2].shape[0]
positive_female = cardio_positive[cardio_positive['gender'] == 1].shape[0]

total_male = df[df['gender'] == 2].shape[0]
total_female = df[df['gender'] == 1].shape[0]

male_percentage = (positive_male / total_male) * 100
female_percentage = (positive_female / total_female) * 100

print("Percentage of males with cardiovascular disease:", round(male_percentage, 2), "%")
print("Percentage of females with cardiovascular disease:", round(female_percentage, 2), "%")
```
