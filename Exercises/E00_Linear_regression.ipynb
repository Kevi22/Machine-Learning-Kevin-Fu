```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
x_values = np.random.normal(100,100,400)
x_values = np.abs(x_values)
noise = np.random.normal(0,50,400)
y_values = 2 * x_values + 25 + noise 

plt.figure(figsize=(10,7))
plt.scatter(x_values, y_values,color ='red')
plt.ylabel('SEK/month')
plt.xlabel('Call minutes')
plt.axhline(0,color='orange',linestyle='--') 
plt.axvline(300,color='green',linestyle='--')
plt.show()
```

```python
np.random.seed(42)
x_values = np.random.normal(100,100,400)
x_values = np.abs(x_values)
x_values= x_values[(x_values>=0) & (x_values<=300)]
noise = np.random.normal(0, 50, len(x_values))
y_values = 2 * x_values + 25 + noise 
b0 = 25
b1 = 2
x_range = np.linspace(0,300,100)
y_true = b0 + b1 * x_range
plt.figure(figsize=(10,7))
plt.scatter(x_values,y_values)

plt.plot(x_range, y_true,color ='red')
plt.ylabel('SEK/month')
plt.xlabel('Call minutes')
plt.show()
```
